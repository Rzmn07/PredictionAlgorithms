```python
# Librerias
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Cargar Dataset
df= pd.read_csv("/content/contracts.csv")
# Descripcion de los datos
df.describe()
```

```python
df.head()
```

```python
# Limpieza de dataset
df.dropna(how="any", inplace=True)
```

```python
# Cambiar variables categoricas a numericas (si se requiere)
df = pd.get_dummies(df, columns=[''], drop_first=True)
df.head()
```

```python
df['single_bidder'] = df['single_bidder'].astype(np.int64)
```

```python
# Seleccionar una submuestra aleatoria 
sa_df= df.sample(frac=0.2,random_state=42)
```

```python
# Seleccion de variables para el modelo
data= sa_df[['est_value','contract_value','n_bids','single_bidder']]
```

```python
# Seleccion de variables para el modelo
data= df[['est_value','contract_value','n_bids','single_bidder']]
```

```python
# Informacion del dataset
data.info()
```

```python
# X son nuestras variables independientes
X = data.drop(["single_bidder"],axis = 1)
```

```python
X.info()
```

```python
y= data["single_bidder"]
```

```python
y
```

```python
# SMOTE
from imblearn.over_sampling import SMOTE

smote= SMOTE()

# Encajar predictor con variable objetivo
x_smote, y_smote = smote.fit_resample(X, y)
```

```python
y
```

```python
y_smote
```

```python
df['single_bidder'] = df['single_bidder'].astype(np.int64)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Definimos modelo RF
RF_model = RandomForestClassifier(n_estimators = 7, random_state = 2016, min_samples_leaf = 8,)
```

```python
# Definimos modelo GB
from sklearn.ensemble import GradientBoostingRegressor

gbr= GradientBoostingRegressor(max_depth=2, n_estimators=6, learning_rate=0.5, min_samples_leaf = 8,)
```

```python
# Definimos modelo LR
from sklearn.linear_model import LogisticRegression

LR_model = LogisticRegression(max_iter = 6000, n_jobs=-1, C=0.5)
```

```python
# Dividimos los datos en entrenamiento y prueba
##### PARA CADA MODELO ######
from sklearn.model_selection import train_test_split

# Dividimos 80% de datos para entrenamiento, 20% de datos para test
X_train, X_test, y_train, y_test = train_test_split(x_smote, y_smote,test_size=0.2)
```

```python
RF_model.fit(X_train, y_train)
```

```python
# Entrenamiento modelo GB
GBR_model = gbr.fit(X_train, y_train)
```

```python
LR_model.fit(X_train, y_train)
```

```python
# Accuracy promedio
RF_model.score(X_test, y_test)
```

```python
# Accuracy promedio
GBR_model.score(X_test, y_test)
```

```python
# Accuracy promedio
LR_model.score(X_test, y_test)
```
